```python
import pandas as pd

df = pd.read_csv("../data/processed/forecasts/nhs_sunderland_forecast.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

# Convert date properly
df['date'] = pd.to_datetime(df['date'])

# Plot
plt.figure()

plt.plot(df['date'], df['quantity'], label='Actual')
plt.plot(df['date'], df['forecast'], label='Forecast')

plt.legend()
plt.title("Pharmacy Demand Forecast - Sunderland")
plt.xticks(rotation=45)

plt.show()
```
